Q.1 To create a list of N points between two pairs of integers A(x1,y1) and B(x2,y2)

```python
import numpy as np

lst = np.linspace((3,5),(30,40),100, endpoint=False)[1:]
lst
```

```python
A = (4,5)
B = (30,40)
arr = []
for i in range(101):
  x = A[0] + round((i * (B[0] - A[0])) / 101, 2)
  y = A[1] + round((i * (B[1] - A[1])) / 101, 2)
  arr.append([x,y])

len(arr), arr[1:]
```
