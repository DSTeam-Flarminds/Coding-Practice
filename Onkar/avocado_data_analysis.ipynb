## Region wise analysis of features

```python
# Group dataframe based on the region
region_group = avacado_df.groupby('region')
```

```python
avacado_df.tail()
```

```python
# Region wise sum of the Total Volume sorted descending of AvergaePrice 
sum_of_total_volume = region_group['Total Volume'].sum().reset_index().sort_values(by = 'Total Volume', ascending = False)
display(sum_of_total_volume.head())
display(sum_of_total_volume.tail())
```
