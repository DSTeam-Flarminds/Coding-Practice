##### Question 1: Write a program that takes two pairs of integers of form A(x1,y1), B(x2,y2) and an integer N. The program should create a list/array of N points between A and B, excluding both of these. And return this array/list of 100 points.

A(x1,y1),B(x2,y2) might be point of a straight line ..which has infinie points between them.

```python
def pointsOfLine(A,B,N):
    y=B[1]-A[1]
    x= B[0]-A[0]
    slope = y/x
    print("slope of line:",slope)
    #for N point located between them 
    d1=(max(A[0],B[0])-min(A[0],B[0]))/(N+1)
    #y=m*x+c straight line equation
    c=B[1]-(slope*B[0])
    points=[]
    x_axis=min(A[0],B[0])+d1
    
    while(x_axis<max(A[0],B[0])):
        #equation y=m*x+c
        curr=[]
        y_axis=(slope*x_axis)+c
        curr.append(x_axis)
        curr.append(y_axis)
        points.append(curr)
        x_axis+=d1
    print(len(points))    
    return points
    
    
    
```

```python
pointsOfLine([0,0],[2,3],100)
```

```python
pointsOfLine([4,1],[2,3],100)
```
